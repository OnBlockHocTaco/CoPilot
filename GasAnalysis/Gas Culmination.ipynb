```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
xl = pd.ExcelFile("SELF-19712 - Monthly pull by trim.xlsx")
df = xl.parse("Sheet1")
Dates = ["2020-05-01", "2020-06-01", "2020-07-01", "2020-08-01", "2020-09-01", "2020-10-01", "2020-11-01",
        "2020-12-01", "2021-01-01", "2021-02-01", "2021-03-01", "2021-04-01"]
CompleteStandardized = []
Month = ["May", "June", "July", "August", "September", "October", "November", "December", "January", "February", "March", "April"]
MonthData = {'Date': Dates,
        'Word': Month}
Months = pd.DataFrame(MonthData)  
CompleteStandardized = []
```

```python
#Chevrolet Equinox

Equinox2016 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2016) & (df["DivisionName"] == "Chevrolet") & (df["ModelName"] == "Equinox") & (df["Trim"] == "LT") & (df['StockType'] == "used")]
Equinox2017 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2017) & (df["DivisionName"] == "Chevrolet") & (df["ModelName"] == "Equinox") & (df["Trim"] == "LT") & (df['StockType'] == "used")]
Equinox2018 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2018) & (df["DivisionName"] == "Chevrolet") & (df["ModelName"] == "Equinox") & (df["Trim"] == "LT") & (df['StockType'] == "used")]
Equinox2019 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2019) & (df["DivisionName"] == "Chevrolet") & (df["ModelName"] == "Equinox") & (df["Trim"] == "LT") & (df['StockType'] == "used")]
l = []
for D in Dates:
    num2016 = Equinox2016.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2017 = Equinox2017.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2018 = Equinox2018.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2019 = Equinox2019.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    l += [np.average([num2016, num2017, num2018, num2019])]

data = {'Date': Dates,
        'Price': l}
Equinox = pd.DataFrame(data)  
maximum = max(l)
minimum = min(l)
standard = []
for val in l:
    standardized = (val - minimum) / (maximum - minimum)
    standard += [standardized]
data = {'Date': Dates,
        'Price': standard}
Equinox = pd.DataFrame(data)  
CompleteStandardized.append(standard)  
Equinox.plot.line(x='Date', y='Price')  



```

```python
#Honda CR-V
CRV2017 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2017) & (df["DivisionName"] == "Honda") & (df["ModelName"] == "CR-V") & (df["Trim"] == "EX-L") & (df["StockType"] == "used")]
CRV2018 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2018) & (df["DivisionName"] == "Honda") & (df["ModelName"] == "CR-V") & (df["Trim"] == "EX") & (df["StockType"] == "used")]
CRV2019 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2019) & (df["DivisionName"] == "Honda") & (df["ModelName"] == "CR-V") & (df["Trim"] == "EX-L") & (df["StockType"] == "used")]
l = []
last2017 = 0
last2018 = 0
last2019 = 0
for D in Dates:
    num2017 = 0;
    num2018 = 0;
    num2019 = 0;
    if D == "2020-09-01":
        num2017 += (last2017 + CRV2017.loc[(df['ReportDate'] == "2020-10-01")]['avgPrice'].iloc[0]) / 2
        num2018 += (last2018 + CRV2018.loc[(df['ReportDate'] == "2020-10-01")]['avgPrice'].iloc[0]) / 2
        num2019 += (last2019 + CRV2019.loc[(df['ReportDate'] == "2020-10-01")]['avgPrice'].iloc[0]) / 2
    else:
        num2017 = CRV2017.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
        last2017 = num2017
        num2018 = CRV2018.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
        last2018 = num2018
        num2019 = CRV2019.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
        last2019 = num2019
    l += [np.average([num2017, num2018, num2019])]

data = {'Date': Dates,
        'Price': l}
CRV = pd.DataFrame(data)  

maximum = max(l)
minimum = min(l)
standard = []
for val in l:
    standardized = (val - minimum) / (maximum - minimum)
    standard += [standardized]
data = {'Date': Dates,
        'Price': standard}
CRV = pd.DataFrame(data)  
CompleteStandardized.append(standard)  
CRV.plot.line(x='Date', y='Price')  


```

```python
gas = []
for i in range(len(Dates)):
    gassum = 0
    standardLength = len(CompleteStandardized)
    for standards in CompleteStandardized:
        gassum += standards[i]
    gas.append(gassum / standardLength)    
data = {'Date': Dates,
        'Price': gas}
GasVehicles = pd.DataFrame(data)  
GasVehicles.plot.line(x='Date', y='Price') 
```

```python
plt.figure(figsize=(14,6))
plt.title("Average Standardized Selling Price of Gasoline Vehicles")
lp = sns.lineplot(x = Dates, y = GasVehicles["Price"])
lp.set(xlabel='Date', ylabel='Average Standardized Price by Month')
setter = lp.set_xticklabels(Months['Word'])

```

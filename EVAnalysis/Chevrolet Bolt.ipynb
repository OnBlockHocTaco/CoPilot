```python
import pandas as pd
import numpy as np
xl = pd.ExcelFile("SELF-19712 - Monthly pull by trim.xlsx")
df = xl.parse("Sheet1")
Dates = ["2020-05-01", "2020-06-01", "2020-07-01", "2020-08-01", "2020-09-01", "2020-10-01", "2020-11-01",
        "2020-12-01", "2021-01-01", "2021-02-01", "2021-03-01", "2021-04-01"]
```

```python
Bolt2017TrimCheck = df.loc[(df['ReportDate'] == "2020-01-01") & (df['ModelYear'] == 2017) & (df["DivisionName"] == "Chevrolet") & (df["ModelName"] == "Bolt EV") & (df["StockType"] == "used")]
Bolt2017TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
Bolt2017 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2017) & (df["DivisionName"] == "Chevrolet") & (df["ModelName"] == "Bolt EV") & (df["StockType"] == "used") & (df["Trim"] == "LT")]
Bolt2017.plot.line(x='ReportDate', y='avgPrice')
```

```python
Bolt2018TrimCheck = df.loc[(df['ReportDate'] == "2020-05-01") & (df['ModelYear'] == 2018) & (df["DivisionName"] == "Chevrolet") & (df["ModelName"] == "Bolt EV") & (df["StockType"] == "used")]
Bolt2018TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
Bolt2018 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2018) & (df["DivisionName"] == "Chevrolet") & (df["ModelName"] == "Bolt EV") & (df["StockType"] == "used") & (df["Trim"] == "LT")]
Bolt2018.plot.line(x='ReportDate', y='avgPrice')
```

```python
Bolt2019TrimCheck = df.loc[(df['ReportDate'] == "2020-08-01") & (df['ModelYear'] == 2019) & (df["DivisionName"] == "Chevrolet") & (df["ModelName"] == "Bolt EV") & (df["StockType"] == "used")]
Bolt2019TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
Bolt2019 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2019) & (df["DivisionName"] == "Chevrolet") & (df["ModelName"] == "Bolt EV") & (df["StockType"] == "used") & (df["Trim"] == "LT")]
Bolt2019.plot.line(x='ReportDate', y='avgPrice')
```

```python
l = []
for D in Dates:
    num2017 = Bolt2017.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2018 = Bolt2018.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2019 = Bolt2019.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    l += [np.average([num2017, num2018, num2019])]

data = {'Date': Dates,
        'Price': l}
Bolt = pd.DataFrame(data)  

Bolt.plot.line(x='Date', y='Price')
```

```python
maximum = max(l)
minimum = min(l)
standard = []
for val in l:
    standardized = (val - minimum) / (maximum - minimum)
    standard += [standardized]
data = {'Date': Dates,
        'Price': standard}
Bolt = pd.DataFrame(data)  

Bolt.plot.line(x='Date', y='Price')  
```

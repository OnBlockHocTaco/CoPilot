```python
import pandas as pd
import numpy as np
xl = pd.ExcelFile("SELF-19712 - Monthly pull by trim.xlsx")
df = xl.parse("Sheet1")
Dates = ["2020-05-01", "2020-06-01", "2020-07-01", "2020-08-01", "2020-09-01", "2020-10-01", "2020-11-01",
        "2020-12-01", "2021-01-01", "2021-02-01", "2021-03-01", "2021-04-01"]

```

```python
ModelS2015TrimCheck = df.loc[(df['ReportDate'] == "2020-05-01") & (df['ModelYear'] == 2015) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model S")]
ModelS2015TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
ModelS2015 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2015) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model S") & (df["Trim"] == "85D")]
ModelS2015.plot.line(x='ReportDate', y='avgPrice')
```

```python
ModelS2016TrimCheck = df.loc[(df['ReportDate'] == "2020-05-01") & (df['ModelYear'] == 2016) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model S")]
ModelS2016TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
ModelS2016 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2016) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model S") & (df["Trim"] == "90D")]
ModelS2016.plot.line(x='ReportDate', y='avgPrice')
```

```python
ModelS2017TrimCheck = df.loc[(df['ReportDate'] == "2020-05-01") & (df['ModelYear'] == 2017) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model S")]
ModelS2017TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
ModelS2017 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2017) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model S") & (df["Trim"] == "100D")]
ModelS2017.plot.line(x='ReportDate', y='avgPrice')
```

```python
ModelS2018TrimCheck = df.loc[(df['ReportDate'] == "2020-05-01") & (df['ModelYear'] == 2018) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model S")]
ModelS2018TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
ModelS2018 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2018) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model S") & (df["Trim"] == "75D")]
ModelS2018.plot.line(x='ReportDate', y='avgPrice')
```

```python
import numpy as np
l = []
for D in Dates:
    num2015 = ModelS2015.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2016 = ModelS2016.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2017 = ModelS2017.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2018 = ModelS2018.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    l += [np.average([num2015, num2016, num2017, num2018])]

data = {'Date': Dates,
        'Price': l}
ModelS = pd.DataFrame(data)  

ModelS.plot.line(x='Date', y='Price')
```

```python
maximum = max(l)
minimum = min(l)
standard = []
for val in l:
    standardized = (val - minimum) / (maximum - minimum)
    standard += [standardized]
data = {'Date': Dates,
        'Price': standard}
ModelS = pd.DataFrame(data)  

ModelS.plot.line(x='Date', y='Price')  
```

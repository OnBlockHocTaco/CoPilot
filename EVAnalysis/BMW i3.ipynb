```python
import numpy as np
l = []
for D in Dates:
    num2015 = i32015.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2016 = i32016.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2017 = i32017.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2018 = i32018.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    l += [np.average([num2015, num2016, num2017, num2018])]

data = {'Date': Dates,
        'Price': l}
i3 = pd.DataFrame(data)  

i3.plot.line(x='Date', y='Price')
```

```python
maximum = max(l)
minimum = min(l)
standard = []
for val in l:
    standardized = (val - minimum) / (maximum - minimum)
    standard += [standardized]
data = {'Date': Dates,
        'Price': standard}
i3 = pd.DataFrame(data)  

i3.plot.line(x='Date', y='Price')  
```

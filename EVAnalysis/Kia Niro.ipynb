```python
Niro2020 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2020) & (df["DivisionName"] == "Kia") & (df["ModelName"] == "Niro") & (df["Trim"] == "LX") & (df['StockType'] == "used")]
Niro2020.plot.line(x='ReportDate', y='avgPrice')
```

```python
l = []
for D in Dates:
    num2017 = Niro2017.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2018 = Niro2018.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2019 = Niro2019.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2020 = 0
    if D == "2020-05-01" or D == "2020-06-01":
        second = Niro2020.loc[(df['ReportDate'] == "2020-08-01")]['avgPrice'].iloc[0]
        num2020 = second
    else:
        num2020 = Niro2020.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    l += [np.average([num2017, num2018, num2019, num2020])]

data = {'Date': Dates,
        'Price': l}
Niro = pd.DataFrame(data)  

Niro.plot.line(x='Date', y='Price')
```

```python
maximum = max(l)
minimum = min(l)
standard = []
for val in l:
    standardized = (val - minimum) / (maximum - minimum)
    standard += [standardized]
data = {'Date': Dates,
        'Price': standard}
Niro = pd.DataFrame(data)  

Niro.plot.line(x='Date', y='Price')  
```

```python
import pandas as pd
import numpy as np
xl = pd.ExcelFile("SELF-19712 - Monthly pull by trim.xlsx")
df = xl.parse("Sheet1")
Dates = ["2020-05-01", "2020-06-01", "2020-07-01", "2020-08-01", "2020-09-01", "2020-10-01", "2020-11-01",
        "2020-12-01", "2021-01-01", "2021-02-01", "2021-03-01", "2021-04-01"]


```

```python
AudiModels = []
eTronCheck = df.loc[(df['ReportDate'] == "2020-05-01") & (df['ModelYear'] == 2017) & (df["DivisionName"] == "Audi") & (~df["ModelName"].isin(AudiModels))]
eTronCheck.head(50)
```

```python
eTron2016TrimCheck = df.loc[(df['ReportDate'] == "2020-05-01") & (df['ModelYear'] == 2016) & (df["DivisionName"] == "Audi") & (df["ModelName"] == "A3 e-tron")]
eTron2016TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
eTron2016 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2016) & (df["DivisionName"] == "Audi") & (df["ModelName"] == "A3 e-tron") & (df["Trim"] == "Premium") & (df['StockType'] == "used")]
eTron2016.plot.line(x='ReportDate', y='avgPrice')
```

```python
eTron2017TrimCheck = df.loc[(df['ReportDate'] == "2020-05-01") & (df['ModelYear'] == 2017) & (df["DivisionName"] == "Audi") & (df["ModelName"] == "A3 Sportback e-tron")]
eTron2017TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
eTron2017 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2017) & (df["DivisionName"] == "Audi") & (df["ModelName"] == "A3 Sportback e-tron") & (df["Trim"] == "Premium") & (df['StockType'] == "used")]
eTron2017.plot.line(x='ReportDate', y='avgPrice')
```

```python
eTron2018TrimCheck = df.loc[(df['ReportDate'] == "2020-05-01") & (df['ModelYear'] == 2018) & (df["DivisionName"] == "Audi") & (df["ModelName"] == "A3 Sportback e-tron")]
eTron2018TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
eTron2018 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2018) & (df["DivisionName"] == "Audi") & (df["ModelName"] == "A3 Sportback e-tron") & (df["Trim"] == "Premium") & (df['StockType'] == "used")]
eTron2018.plot.line(x='ReportDate', y='avgPrice')
```

```python
import numpy as np
l = []
for D in Dates:
    num2016 = eTron2016.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2017 = eTron2017.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2018 = eTron2018.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    l += [np.average([num2016, num2017, num2018])]

data = {'Date': Dates,
        'Price': l}
eTron = pd.DataFrame(data)  

eTron.plot.line(x='Date', y='Price')
```

```python
maximum = max(l)
minimum = min(l)
standard = []
for val in l:
    standardized = (val - minimum) / (maximum - minimum)
    standard += [standardized]
data = {'Date': Dates,
        'Price': standard}
eTron = pd.DataFrame(data)  

eTron.plot.line(x='Date', y='Price')  
```

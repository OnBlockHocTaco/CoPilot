```python
eGolf2018 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2018) & (df["DivisionName"] == "Volkswagen") & (df["ModelName"] == "e-Golf") & (df["Trim"] == "SE") & (df['StockType'] == "used")]
eGolf2018.plot.line(x='ReportDate', y='avgPrice')
```

```python
import numpy as np
l = []
for D in Dates:
    num2015 = eGolf2015.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2016 = eGolf2016.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2017 = eGolf2017.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2018 = eGolf2018.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    l += [np.average([num2015, num2016, num2017, num2018])]

data = {'Date': Dates,
        'Price': l}
eGolf = pd.DataFrame(data)  

eGolf.plot.line(x='Date', y='Price')
```

```python
maximum = max(l)
minimum = min(l)
standard = []
for val in l:
    standardized = (val - minimum) / (maximum - minimum)
    standard += [standardized]
data = {'Date': Dates,
        'Price': standard}
eGolf = pd.DataFrame(data)  

eGolf.plot.line(x='Date', y='Price')  
```

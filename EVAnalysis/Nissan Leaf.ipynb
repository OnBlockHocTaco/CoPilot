```python
import numpy as np
l = []
for D in Dates:
    num2015 = Leaf2015.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2016 = Leaf2016.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2017 = Leaf2017.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2018 = Leaf2018.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    l += [np.average([num2015, num2016, num2017, num2018])]

data = {'Date': Dates,
        'Price': l}
Leaf = pd.DataFrame(data)  

Leaf.plot.line(x='Date', y='Price')
```

```python
maximum = max(l)
minimum = min(l)
standard = []
for val in l:
    standardized = (val - minimum) / (maximum - minimum)
    standard += [standardized]
data = {'Date': Dates,
        'Price': standard}
Leaf = pd.DataFrame(data)  

Leaf.plot.line(x='Date', y='Price')  
```

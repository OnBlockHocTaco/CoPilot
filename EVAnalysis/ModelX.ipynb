```python
import pandas as pd
import numpy as np
xl = pd.ExcelFile("SELF-19712 - Monthly pull by trim.xlsx")
df = xl.parse("Sheet1")
Dates = ["2020-05-01", "2020-06-01", "2020-07-01", "2020-08-01", "2020-09-01", "2020-10-01", "2020-11-01",
        "2020-12-01", "2021-01-01", "2021-02-01", "2021-03-01", "2021-04-01"]

```

```python
ModelX2016TrimCheck = df.loc[(df['ReportDate'] == "2020-05-01") & (df['ModelYear'] == 2016) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model X")]
ModelX2016TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
ModelX2016 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2016) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model X") & (df["Trim"] == "90D") & (df['StockType'] == "used")]
ModelX2016.plot.line(x='ReportDate', y='avgPrice')
```

```python
ModelX2017TrimCheck = df.loc[(df['ReportDate'] == "2020-05-01") & (df['ModelYear'] == 2017) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model X")]
ModelX2017TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
ModelX2017 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2017) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model X") & (df["Trim"] == "75D") & (df['StockType'] == "used")]
ModelX2017.plot.line(x='ReportDate', y='avgPrice')
```

```python
ModelX2018TrimCheck = df.loc[(df['ReportDate'] == "2020-05-01") & (df['ModelYear'] == 2018) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model X")]
ModelX2018TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
ModelX2018 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2018) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model X") & (df["Trim"] == "75D") & (df['StockType'] == "used")]
ModelX2018.plot.line(x='ReportDate', y='avgPrice')
```

```python
ModelX2019TrimCheck = df.loc[(df['ReportDate'] == "2020-05-01") & (df['ModelYear'] == 2019) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model X")]
ModelX2019TrimCheck.plot.bar(x='Trim', y='vehicleCount')
```

```python
ModelX2019 = df.loc[(df['ReportDate'].isin(Dates)) & (df['ModelYear'] == 2019) & (df["DivisionName"] == "Tesla") & (df["ModelName"] == "Model X") & (df["Trim"] == "Performance") & (df['StockType'] == "used")]
ModelX2019.plot.line(x='ReportDate', y='avgPrice')
```

```python
import numpy as np
l = []
for D in Dates:
    num2016 = ModelX2016.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2017 = ModelX2017.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2018 = ModelX2018.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    num2019 = ModelX2019.loc[(df['ReportDate'] == D)]['avgPrice'].iloc[0]
    l += [np.average([num2016, num2017, num2018, num2019])]

data = {'Date': Dates,
        'Price': l}
ModelX = pd.DataFrame(data)  

ModelX.plot.line(x='Date', y='Price')
```

```python
maximum = max(l)
minimum = min(l)
standard = []
for val in l:
    standardized = (val - minimum) / (maximum - minimum)
    standard += [standardized]
data = {'Date': Dates,
        'Price': standard}
ModelX = pd.DataFrame(data)  

ModelX.plot.line(x='Date', y='Price')  
```
